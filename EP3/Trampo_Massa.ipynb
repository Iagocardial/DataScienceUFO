```python
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("Ep3-Masanori-Final.csv", low_memory=False, encoding="latin-1", delimiter=";", decimal=',')
df.head()
```

```python
plt.axis('equal')
plt.pie(df['país'].value_counts().values, labels = df['país'].value_counts().index, shadow=False, explode=(0, 0, 0, 0,0.2),
startangle=160,autopct='%.2f')
plt.suptitle("Países com mais avistamentos de UFO")
plt.tight_layout()
plt.show()
```

```python
plt.axis('equal')
plt.pie(df['shape'].value_counts().values, labels = df['shape'].value_counts().index, shadow=False,
startangle=160,autopct='%.2f')
plt.suptitle("Formas mais vistas")
plt.tight_layout()
plt.show()
```

```python
df1 = df.replace(np.nan, '', regex=True)
scatter_plot = plt.scatter(df1['duration_segundos'], df1['país'], alpha=0.22)
plt.show()
```

```python
dt_f = df.groupby(['país']).apply(lambda x: x['duration_segundos'].mean())
print("Média de segundos para cada {}".format(dt_f))
```
